```python

import json
# DSLs = ['ASTMatcher', 'Flight', 'sklearn', 'TextEditing']
# DSLs = ['TextEditing']
# DSLs = ['ASTMatcher']
DSLs = ['Flight']
# DSLs = ['sklearn']
for DSL in DSLs:
    length = len(open(f'/home/scruple/NLP/HISyn/Documentation/{DSL}/code_new.txt').readlines())
    with open(f'/home/scruple/NLP/HISyn/Documentation/{DSL}/code_new.txt', 'r') as codes:
        with open(f'/home/scruple/NLP/HISyn/Documentation/{DSL}/text_new.txt', 'r') as texts:
            
            train_size = int(0.7*length)
            valid_size = int(0.2*length)
            print(train_size, valid_size)
            cnt = 0
            print('train')
            with open('/home/scruple/NLP/base/datasets/CERT/NLTK/data/train/try/output.jsonl', 'w') as outfile:
                for text, code in zip(texts, codes):
                    entry = {'content': "\nLIBRARY:\n" + DSL + '\n' + "\nQUESTION:\n" + text + "\n" + "\nANSWER:\n"  + code}
                    print(entry['content'])
                    json.dump(entry, outfile)
                    outfile.write('\n')
                    cnt+=1
                    if cnt>=train_size:
                        break
            cnt = 0
            print('valid')
            with open('/home/scruple/NLP/base/datasets/CERT/NLTK/data/valid/try/output.jsonl', 'w') as outfile:
                for text, code in zip(texts, codes):
                    entry = {'content': "\nLIBRARY:\n" + DSL + '\n' + "\nQUESTION:\n" + text + "\n" + "\nANSWER:\n"  + code}
                    print(entry['content'])
                    json.dump(entry, outfile)
                    outfile.write('\n')
                    cnt+=1
                    if cnt>=valid_size:
                        break

                

```
